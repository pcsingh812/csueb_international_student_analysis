```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import textwrap

# Count occurrences of importance levels
importance_counts = {
    'Factor': ['Tuition fees', 'Coursework & Flexible Schedules', 
               'Location and Infrastructure', 'School Ranking'],
    'Important': [
        (df['Tuition fees'] == 'Important').sum(),
        (df['Coursework & Flexible Schedules'] == 'Important').sum(),
        (df['Location and Infrastructure'] == 'Important').sum(),
        (df['School Ranking'] == 'Important').sum()
    ],
    'Moderately important': [
        (df['Tuition fees'] == 'Moderately important').sum(),
        (df['Coursework & Flexible Schedules'] == 'Moderately important').sum(),
        (df['Location and Infrastructure'] == 'Moderately important').sum(),
        (df['School Ranking'] == 'Moderately important').sum()
    ],
    'Not important': [
        (df['Tuition fees'] == 'Not important').sum(),
        (df['Coursework & Flexible Schedules'] == 'Not important').sum(),
        (df['Location and Infrastructure'] == 'Not important').sum(),
        (df['School Ranking'] == 'Not important').sum()
    ]
}

# Convert the dictionary to a DataFrame for plotting
df_counts = pd.DataFrame(importance_counts)

# Set the seaborn style for a clean look
sns.set(style="whitegrid")

# Define a pastel color palette
pastel_colors = sns.color_palette("pastel", 3)  # Soft pastel colors

# Create the stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the stacked bars
ax.bar(df_counts['Factor'], df_counts['Important'], label='Important', color=pastel_colors[0])
ax.bar(df_counts['Factor'], df_counts['Moderately important'], 
       bottom=df_counts['Important'], label='Moderately important', color=pastel_colors[1])
ax.bar(df_counts['Factor'], df_counts['Not important'], 
       bottom=df_counts['Important'] + df_counts['Moderately important'], label='Not important', color=pastel_colors[2])

# Add counts on top of each bar segment
for index, row in df_counts.iterrows():
    # Important
    ax.text(index, row['Important'] / 2, str(row['Important']), ha='center', va='center', color='black', fontweight='bold')
    # Moderately Important
    ax.text(index, row['Important'] + row['Moderately important'] / 2, str(row['Moderately important']), ha='center', va='center', color='black', fontweight='bold')
    # Not Important
    ax.text(index, row['Important'] + row['Moderately important'] + row['Not important'] / 2, str(row['Not important']), ha='center', va='center', color='black', fontweight='bold')

# Add titles and labels with bold font
ax.set_ylabel('Count', fontsize=15, fontweight='bold')
ax.set_title('Importance Levels of Factors', fontsize=17, fontweight='bold')

# Move the legend outside the plot
ax.legend(title="Importance Level", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=15)

# Rotate x-axis labels for readability and wrap text with bold font
ax.set_xticklabels([textwrap.fill(label, 10) for label in df_counts['Factor']], fontsize=15, fontweight='bold')

# Bold y-axis tick marks
plt.yticks(fontsize=16, fontweight='bold')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import textwrap

# Count occurrences of support levels and rename factors
support_counts = {
    'Factor1': ['Career Planning', 
                'Basic needs', 
                'Financial aid', 
                'Safety', 
                'DEI'],
    'Very supported': [
        (df['Career and post-completion planning'] == 'Very supported').sum(),
        (df['Basic needs (housing, food, medicine, mental health)'] == 'Very supported').sum(),
        (df['Financial aid'] == 'Very supported').sum(),
        (df['Safety'] == 'Very supported').sum(),
        (df['DEI(Diversity, Equity and Inclusion)'] == 'Very supported').sum()
    ],
    'Somewhat supported': [
        (df['Career and post-completion planning'] == 'Somewhat supported').sum(),
        (df['Basic needs (housing, food, medicine, mental health)'] == 'Somewhat supported').sum(),
        (df['Financial aid'] == 'Somewhat supported').sum(),
        (df['Safety'] == 'Somewhat supported').sum(),
        (df['DEI(Diversity, Equity and Inclusion)'] == 'Somewhat supported').sum()
    ],
    'Not supported': [
        (df['Career and post-completion planning'] == 'Not supported').sum(),
        (df['Basic needs (housing, food, medicine, mental health)'] == 'Not supported').sum(),
        (df['Financial aid'] == 'Not supported').sum(),
        (df['Safety'] == 'Not supported').sum(),
        (df['DEI(Diversity, Equity and Inclusion)'] == 'Not supported').sum()
    ]
}

# Convert the dictionary to a DataFrame for plotting
df_counts = pd.DataFrame(support_counts)

# Set the Seaborn style
sns.set(style="whitegrid")

# Define a pastel color palette
pastel_colors = sns.color_palette("pastel", 3)

# Create the stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the stacked bars
ax.bar(df_counts['Factor1'], df_counts['Very supported'], label='Very supported', color=pastel_colors[0])
ax.bar(df_counts['Factor1'], df_counts['Somewhat supported'], 
       bottom=df_counts['Very supported'], label='Somewhat supported', color=pastel_colors[1])
ax.bar(df_counts['Factor1'], df_counts['Not supported'], 
       bottom=df_counts['Very supported'] + df_counts['Somewhat supported'], label='Not supported', color=pastel_colors[2])

# Add labels and title with larger fonts for clarity
ax.set_ylabel('Count', fontsize=15, fontweight='bold')
ax.set_title('Support Levels for Various Factors', fontsize=17, fontweight='bold')

# Move the legend outside the plot
ax.legend(title="Support Level", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

# Set x-tick labels and alignment
wrapped_labels = [textwrap.fill(label, width=10) for label in df_counts['Factor1']]
ax.set_xticklabels(wrapped_labels, rotation=0, ha="center", fontsize=12, fontweight='bold')

# Set y-ticks fontsize
plt.yticks(fontsize=15, fontweight='bold')

# Add counts on top of the bars
for i in range(len(df_counts)):
    # Get the counts for each level of support
    very_supported = df_counts['Very supported'][i]
    somewhat_supported = df_counts['Somewhat supported'][i]
    not_supported = df_counts['Not supported'][i]
    
    # Add text labels for each category
    if very_supported > 0:
        ax.text(i, very_supported / 2, str(very_supported), ha='center', va='center', fontsize=9, color='black', fontweight='bold')
    if somewhat_supported > 0:
        ax.text(i, very_supported + somewhat_supported / 2, str(somewhat_supported), ha='center', va='center', fontsize=9, color='black', fontweight='bold')
    if not_supported > 0:
        ax.text(i, very_supported + somewhat_supported + not_supported / 2, str(not_supported), ha='center', va='center', fontsize=9, color='black', fontweight='bold')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import textwrap

# Data for support counts
support_counts = {
    'Factor2': ['When I have a concern or issue, it is easy for me to find the help I need', 
                'College equipped me with technical and people skills to find internships/jobs'],
    'Agree': [
        (df['When I have a concern or issue, it is easy for me to find the help I need'] == 'Agree').sum(),
        (df['College equipped me with technical and people skills to find internships/jobs'] == 'Agree').sum(),
    ],
    'Disagree': [
        (df['When I have a concern or issue, it is easy for me to find the help I need'] == 'Disagree').sum(),
        (df['College equipped me with technical and people skills to find internships/jobs'] == 'Disagree').sum(),
    ]
}

# Convert the dictionary to a DataFrame for plotting
df_counts = pd.DataFrame(support_counts)

# Create a grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Set bar width
bar_width = 0.35

# Set the position of the bars on the x-axis
index = np.arange(len(df_counts['Factor2']))

# Plot Agree bars
bars1 = ax.bar(index, df_counts['Agree'], width=bar_width, label='Agreed', color='teal')

# Plot Disagree bars next to Agree bars
bars2 = ax.bar(index + bar_width, df_counts['Disagree'], width=bar_width, label='Disagreed', color='skyblue')

# Add labels and title with increased font sizes and bold
ax.set_xlabel('Factors', fontsize=15, fontweight='bold')
ax.set_ylabel('Count', fontsize=15, fontweight='bold')
ax.set_title('Agreement Levels on Various Factors', fontsize=17, fontweight='bold')

# Format the x-axis labels for better readability
wrapped_labels = [textwrap.fill(label, 30) for label in df_counts['Factor2']]  # Wrap text for long labels
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(wrapped_labels, rotation=0, ha="center", fontsize=15, fontweight='bold')

# Add a legend with a title
ax.legend(title="Response", fontsize=12, title_fontsize='13', loc='upper right', frameon=False)

# Annotate the bars with the count values
for bar in bars1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.1, int(yval), ha='center', va='bottom', fontsize=15, fontweight='bold')

for bar in bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.1, int(yval), ha='center', va='bottom', fontsize=15, fontweight='bold')

# Adjust layout for better display
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import textwrap

# Data for support counts
support_counts = {
    'Factor3': ['I know how to get approval for reduced course load or withdrawal', 
                'I know the full time enrollment requirement related to my F1/J1 visa',
                'I understand my work I-20/DS-2019 authorization expiration date options and grace period for departure'],
    'Agree': [
        (df['I know how to get approval for reduced course load or withdrawal'] == 'Agree').sum(),
        (df['I know the full time enrollment requirement related to my F1/J1 visa'] == 'Agree').sum(),
        (df['I understand my work I-20/DS-2019 authorization expiration date options and grace period for departure'] == 'Agree').sum()
    ],
    'Disagree': [
        (df['I know how to get approval for reduced course load or withdrawal'] == 'Disagree').sum(),
        (df['I know the full time enrollment requirement related to my F1/J1 visa'] == 'Disagree').sum(),
        (df['I understand my work I-20/DS-2019 authorization expiration date options and grace period for departure'] == 'Disagree').sum()
    ]
}

# Convert the dictionary to a DataFrame for plotting
df_counts = pd.DataFrame(support_counts)

# Set a fancy style for the plot using seaborn
sns.set(style="whitegrid")

# Set up figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Set bar width
bar_width = 0.35

# Set the position of the bars on the x-axis
index = np.arange(len(df_counts['Factor3']))

# Define colors
agree_color = sns.color_palette("Set2")[0]  # Soft green
disagree_color = sns.color_palette("Set2")[1]  # Soft yellow

# Plot Agree bars
bars1 = ax.bar(index, df_counts['Agree'], width=bar_width, label='Agree', color=agree_color)

# Plot Disagree bars next to Agree bars
bars2 = ax.bar(index + bar_width, df_counts['Disagree'], width=bar_width, label='Disagree', color=disagree_color)

# Add labels and title with increased font sizes and bold
ax.set_xlabel('Factors', fontsize=15, fontweight='bold')
ax.set_ylabel('Count', fontsize=15, fontweight='bold')
ax.set_title('Agreement Levels on Visa and Enrollment Understanding', fontsize=17, fontweight='bold')

# Format the x-axis labels for better readability (wrapping long labels)
wrapped_labels = [textwrap.fill(label, 30) for label in df_counts['Factor3']]  # Wrap text for long labels
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(wrapped_labels, rotation=0, ha="center", fontsize=15, fontweight='bold')

# Add a legend with a title
legend = ax.legend(title="Response", fontsize=15, title_fontsize=15, frameon=False)
legend.get_title().set_fontweight('bold')  # Set the legend title to bold

# Annotate the bars with the count values
for bar in bars1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, int(yval), ha='center', va='bottom', fontsize=15, fontweight='bold')

for bar in bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, int(yval), ha='center', va='bottom', fontsize=15, fontweight='bold')

# Adjust layout to avoid cutting off labels
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Count the occurrences of each category
response_counts = df['To what extent have you personally benefited from the current alumni network?'].value_counts()

# Step 2: Calculate percentage
response_percentages = response_counts / response_counts.sum() * 100

# Step 3: Plot pie chart
fig, ax = plt.subplots(figsize=(8, 8))

# Define a pastel color palette
pastel_colors = sns.color_palette("pastel", n_colors=len(response_counts))

# Create the pie chart with increased font size for labels
ax.pie(
    response_percentages,
    labels=response_percentages.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=pastel_colors,
    textprops={'fontsize': 20, 'fontweight': 'bold'}  # Increase font size and make it bold
)

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')

# Add title in bold
plt.title('Extent of Benefit from Alumni Network', fontsize=16, fontweight='bold')

# Show the plot
plt.show()
```
